## Exploracion y entrenamiento de los datos

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pickle
import os

datos_flores = load_iris()
x_caracteristicas = datos_flores.data
y_etiquetas = datos_flores.target

modelo_sencillo = DecisionTreeClassifier(random_state=42)
modelo_sencillo.fit(x_caracteristicas, y_etiquetas)

# ruta para guardar el modelo en la carpeta models
ruta_modelo = os.path.join("..", "models", "modelo_listo.pkl")

with open(ruta_modelo, 'wb') as archivo:
    pickle.dump(modelo_sencillo, archivo)

print("Modelo entrenado y guardado en models/")
```
